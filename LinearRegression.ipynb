```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import sklearn
from sklearn.datasets import load_boston
boston = load_boston()
bos = pd.DataFrame(boston.data)
import seaborn as sns
```

```python
print(boston.DESCR)
```

```python
bos.columns=boston.feature_names
bos.head()
```

```python
bos['price']=boston.target
```

```python
cor_data=bos.corr()
```

```python
cor_data #correlation matrix
```

```python
fig,ax=plt.subplots(figsize=(15,15))
ax=sns.heatmap(cor_data,annot=True,linewidths=0.8) #visualisation through heatmap
```

```python
#from the heatmap we can see the multicollinearity in the input features 
#& lets consider the features with high correlation with the price
```

```python
filter_matrix=pd.DataFrame(cor_data['price'],index=cor_data.columns[:-1],columns=['price'])
```

```python
filter_matrix
```

```python
#Let us consider the features with higher collinearity with the price
best_features=[feature for feature in filter_matrix.index if abs(filter_matrix.loc[feature]['price'])>=0.45 ]
```

```python
best_features
```

```python
#but RM is highly correlated with LSTAT,So lets consider only LSTAT 
```

```python
best_features.remove('RM')
```

```python
best_features
```

```python
data=bos[best_features]
```

```python
data=data.join(bos.price)
```

```python
for i in data.columns:
    print('skewness in {} = {}'.format(i,data[i].skew()))
```

```python
for i in data.columns:
    for j in data.index:
        data.loc[j][i]=np.log(data.loc[j][i])
```

```python
for i in data.columns:
    print('skewness in {} = {}'.format(i,data[i].skew()))
```

```python
x=data.drop('price',axis=1)
```

```python
y=data['price']
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=np.random)
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
pred_Y=lm.predict(X_test)
```

```python
pred_Y.shape,y_test.shape
```

```python
lm.score(X_test,y_test)
```
